### Чистка данных

Заполним пропущенные значения:

`Age` - медианой (а не средним арифметическим например, т.к. оно очень чувствительно к выбросам).

`Fare` - медианой.

### Классификация `RandomForestClassifier`

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score
from sklearn.model_selection import train_test_split

# Разделим данные на признаки (X) и целевую переменную (y)
X = train_df.drop(columns=['Survived'])
y = train_df['Survived']

# Разделим данные на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Инициализация модели RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)

# Обучение модели
rf_model.fit(X_train, y_train)

# Предсказания на тестовой выборке
y_pred_rf = rf_model.predict(X_test)

# Оценка модели с помощью метрик
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)

# Выводим метрики
print(f"""RandomForestClassifier - \
Precision: {precision_rf:.2f}, \
Recall: {recall_rf:.2f}, \
Accuracy: {accuracy_rf:.2f}, \
F1 Score: {f1_rf:.2f}""")
```

Для нашего анализа метрика `accuracy` выше (и равна 0.82), чем на том, что было на семинаре(0.8). А значит, задача выполнена.

### Регрессия `LinearRegression`

В блоке кода ниже мы обучаем модель линейной регрессии для предсказания возраста пассажиров на основе других признаков. Сначала делим данные на признаки и целевую переменную (`Age`), затем разделяем данные на тренировочную и тестовую выборки. После этого стандартизируем признаки для улучшения работы модели и обучаем линейную регрессию. В конце, используя тестовую выборку, получаем предсказания и рассчитываем метрику `RMSE`, которая показывает точность модели — чем меньше значение, тем точнее модель.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Разделим данные на признаки и целевую переменную
age_features = train_df.drop(columns=['Age', 'Survived'])  # исключаем 'Age' и 'Survived'
age_target = train_df['Age']

# Разделим данные на тренировочную и тестовую выборки
age_features_train, age_features_test, age_target_train, age_target_test = train_test_split(
    age_features, age_target, test_size=0.2, random_state=56
)

# Стандартизируем данные
scaler = StandardScaler()
age_features_train_scaled = scaler.fit_transform(age_features_train)
age_features_test_scaled = scaler.transform(age_features_test)

# Обучаем модель линейной регрессии
model = LinearRegression()
model.fit(age_features_train_scaled, age_target_train)

# Прогнозируем на тестовой выборке
age_pred = model.predict(age_features_test_scaled)

# Рассчитываем RMSE до удаления выбросов
rmse = np.sqrt(mean_squared_error(age_target_test, age_pred))
print(f"RMSE: {rmse}")
```

Здесь мы пытаемся найти выбросы в тестовой выборке, которые могут сильно искажать модель. Для этого рассчитываем отклонения между реальными значениями возраста и предсказанными. Если это отклонение больше чем 3 стандартных отклонения (как правило, выбросы расположены далеко от основного распределения), то такие значения считаются выбросами и исключаются.

```python
# Определение выбросов на основе отклонений от предсказанных значений
threshold = 3 * (age_target_test - age_pred).std()
filtered_indices_test = np.abs(age_target_test - age_pred) < threshold

# Фильтрация выбросов
age_features_test_filtered = age_features_test_scaled[filtered_indices_test]
age_target_test_filtered = age_target_test[filtered_indices_test]

# Повторное обучение модели на отфильтрованных данных
model_filtered = LinearRegression()
model_filtered.fit(age_features_train_scaled, age_target_train)

# Прогнозирование на отфильтрованной тестовой выборке
age_pred_filtered = model_filtered.predict(age_features_test_filtered)

# Рассчитываем RMSE без выбросов
rmse_filtered = np.sqrt(mean_squared_error(age_target_test_filtered, age_pred_filtered))
print(f"RMSE без выбросов: {rmse_filtered}")
```

Как видно, мы успешно оптимизировали модель и получили ошибку `RMSE` (10.47) ниже заявленой на семинаре(10.7) . Значит, задача выполнена
