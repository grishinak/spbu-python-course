## Разделим на тестовую и обучающую выборки

```python
from sklearn.model_selection import train_test_split

X = data.drop(columns=['type'])
y = data['type']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
display(X_train.count())
display(X_test.count())
```

 # $\blacksquare$ Используя **подбор гиперпараметров**, **кросс-валидацию** и при необходимости **масштабирование данных**, добиться наилучшего качества предсказания для библиотечных реализаций (например, из **sklearn**) четырёх методов классификации.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

**Масштабируем** данные(`legs` --- не One-Heated).

```python
from sklearn.preprocessing import StandardScaler

# Создаем объект StandardScaler
scaler = StandardScaler()

# Масштабируем данные
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Лучшие параметры будем **подбирать** с помощью след. функции(там же встроена **кросс-валидация**):

```python
def train_and_optimize(model, param_grid, X, y):
    """
    Обучает модель с использованием GridSearchCV и возвращает лучшую модель.
    """
    grid = GridSearchCV(model, param_grid, cv=4, scoring='accuracy', n_jobs=-1) # cv==cross-validation
    grid.fit(X, y)

    print(f"Лучшие параметры: {grid.best_params_}")
    print(f"Лучший результат (accuracy): {grid.best_score_:.4f}")

    return grid.best_estimator_
```

## DecisionTreeClassifier

```python
dt_params = {
    'max_depth': [3, 5, 7, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

dt_model = train_and_optimize(DecisionTreeClassifier(random_state=42), dt_params, X_train_scaled, y_train)
```

## RandomForestClassifier

```python
rf_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf_model = train_and_optimize(RandomForestClassifier(random_state=42), rf_params, X_train_scaled, y_train)
```

## LogisticRegression

```python
lr_params = {
    'penalty': ['l1', 'l2'],
    'C': [0.1, 1, 10],
    'solver': ['liblinear']
}

lr_model = train_and_optimize(LogisticRegression(random_state=42, max_iter=1000), lr_params, X_train_scaled, y_train)
```

## KNeighborsClassifier

```python
knn_params = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}

knn_model = train_and_optimize(KNeighborsClassifier(), knn_params, X_train_scaled, y_train)
```

```python
# Создаем словарь моделей
models = {
    "Decision Tree": dt_model,
    "Random Forest": rf_model,
    "Logistic Regression": lr_model,
    "K-Nearest Neighbors": knn_model
}

# Выполняем предсказание и оцениваем каждую модель
for name, model in models.items():
    y_pred = model.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred)
    print(f"Модель: {name}")
    print(f"Точность на тестовой выборке: {acc:.4f}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("Classification Report:")
    print(classification_report(y_test, y_pred, zero_division=1))
    print("-" * 50)
```

# $\blacksquare$ Сравнить все обученные модели, построить их **confusion matrices**. Сделать выводы о полученных моделях в рамках решения задачи классификации на выбранных данных.

## Сравнение моделей

1. `Decision Tree и Logistic Regression` достигают идеальных результатов (точность = 1), но есть вероятность переобучения для Decision Tree, особенно на малых выборках.
2. `Random Forest` показывает более реалистичную производительность с меньшей точностью, но сохраняет устойчивость и избегает полного переобучения. Однако проблемы с классификацией редких классов (3 и 5) остаются.
3. `KNN`демонстрирует наименьшую точность среди всех моделей, но может быть полезным для сценариев с большей выборкой и оптимизацией параметра kk.
- Модели `Decision Tree и Logistic Regression` идеально справляются с <u>текущими</u> данными, но стоит протестировать их на большей выборке, чтобы проверить их "общность".

## Построение confusion matrix для каждой модели


```python
from sklearn.metrics import confusion_matrix

fig, axes = plt.subplots(2, 2, figsize=(18, 12))
axes = axes.ravel()  # Плоский массив для удобства

for i, (name, model) in enumerate(models.items()):
    # Получаем предсказания
    y_pred = model.predict(X_test_scaled)

    # Строим confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    # Визуализируем с помощью heatmap
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_, ax=axes[i])

    # Заголовок для каждой матрицы
    axes[i].set_title(f'{name} Confusion Matrix')
    axes[i].set_xlabel('Predicted')
    axes[i].set_ylabel('True')

# Подгонка для компоновки
plt.tight_layout()
plt.show()
```

Как было сказано выше, в сравнении, `DT` и `LR` показывают лучшую точность предсказаний, однако для оставшихся моделей есть неточности с определением редких классов 3 и 5(у `RF` ---1 ошибка, у `KNN` --- 2)

Уменьшим размрность и построим графики того, как справились наши модели, по аналогии с тем, как мы делали в работе с ирисами Фишера

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Снижаем размерность с помощью PCA
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_test_scaled)

# Словарь для хранения предсказанных меток всех моделей
y_pred_all = {}

# Получаем предсказания для каждой модели
for name, model in models.items():
    if hasattr(model, "predict"):
        y_pred_all[name] = model.predict(X_test_scaled)

# Построение графиков для каждой модели
plt.figure(figsize=(15, 12))

for i, (name, y_pred) in enumerate(y_pred_all.items(), 1):
    # График с реальными метками
    plt.subplot(4, 2, i * 2 - 1)  # Индекс для первого столбца
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_test, cmap='Set2', edgecolor='k', s=50)
    plt.title(f'PCA with True Labels: {name}')
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.colorbar()

    # График с предсказанными метками
    plt.subplot(4, 2, i * 2)  # Индекс для второго столбца
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_pred, cmap='Set2', edgecolor='k', s=50)
    plt.title(f'PCA with Predicted Labels: {name}')
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.colorbar()

plt.tight_layout()
plt.show()
```

Здесь отчетливо выделяется область 1 класса, 2 и 7 тоже достаточно хорошо определяются, однако при приближении слева к точке (0,0) наблюдается неопределенность в той области.
