## Разделим на тестовую и обучающую выборки

````python
from sklearn.model_selection import train_test_split

X = data.drop(columns=['type'])
y = data['type']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
````

````python
display(X_train.count())
display(X_test.count())
````

 # $\blacksquare$ Используя **подбор гиперпараметров**, **кросс-валидацию** и при необходимости **масштабирование данных**, добиться наилучшего качества предсказания для библиотечных реализаций (например, из **sklearn**) четырёх методов классификации.

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
````

**Масштабируем** данные(`legs` --- не One-Heated).

````python
from sklearn.preprocessing import StandardScaler

# Создаем объект StandardScaler
scaler = StandardScaler()

# Масштабируем данные
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
````

Выведем после масштабирования:

````python
display(X_train_scaled[:3],X_test_scaled[:3])
````

Лучшие параметры будем **подбирать** с помощью след. функции(там же встроена **кросс-валидация**):

````python
def train_and_optimize(model, param_grid, X, y):
    """
    Обучает модель с использованием GridSearchCV и возвращает лучшую модель.
    """
    # заменила на balanced_accuracy, т.к. она учитывает несбалансированные данные, как тут
    grid = GridSearchCV(model, param_grid, cv=4, scoring='balanced_accuracy', n_jobs=-1) # cv==cross-validation
    grid.fit(X, y)

    print(f"Лучшие параметры: {grid.best_params_}")
    print(f"Лучший результат (balanced_accuracy): {grid.best_score_:.4f}")

    return grid.best_estimator_
````

Будем создавать масштабируемые и не масштабируемые модели для сравнения к коментарию ниже для каждой модели

## DecisionTreeClassifier

````python
dt_params = {
    'max_depth': [3, 5, 7, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

dt_model = train_and_optimize(DecisionTreeClassifier(random_state=42), dt_params, X_train_scaled, y_train)
````

````python
dt_no_scaled_model = train_and_optimize(DecisionTreeClassifier(random_state=42), dt_params, X_train, y_train)
````

## RandomForestClassifier

````python
rf_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf_model = train_and_optimize(RandomForestClassifier(random_state=42), rf_params, X_train_scaled, y_train)
````

````python
rf_no_scaled_model = train_and_optimize(RandomForestClassifier(random_state=42), rf_params, X_train, y_train)
````

## LogisticRegression

````python
lr_params = {
    'penalty': ['l1', 'l2'],
    'C': [0.1, 1, 10],
    'solver': ['liblinear']
}

lr_model = train_and_optimize(LogisticRegression(random_state=42, max_iter=1000), lr_params, X_train_scaled, y_train)
````

````python
lr_no_scaled_model = train_and_optimize(LogisticRegression(random_state=42, max_iter=1000), lr_params, X_train, y_train)
````

## KNeighborsClassifier

````python
knn_params = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}

knn_model = train_and_optimize(KNeighborsClassifier(), knn_params, X_train_scaled, y_train)
````

````python
knn_no_scaled_model = train_and_optimize(KNeighborsClassifier(), knn_params, X_train, y_train)
````

````python
# Создаем словарь моделей
models = {
    "Decision Tree": dt_model,
    "Random Forest": rf_model,
    "Logistic Regression": lr_model,
    "K-Nearest Neighbors": knn_model
}
# Создаем словарь не масштабируемых моделей
models_no_scaled = {
    "Decision Tree": dt_no_scaled_model,
    "Random Forest": rf_no_scaled_model,
    "Logistic Regression": lr_no_scaled_model,
    "K-Nearest Neighbors": knn_no_scaled_model
}

# Выполняем предсказание и оцениваем каждую модель
for name, model in models_no_scaled.items():
    y_pred = model.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred)
    print(f"Модель: {name}")
    print(f"balanced_accuracy на тестовой выборке: {acc:.4f}")
    # to understand before visualisation itself
    # print("Confusion Matrix:")
    # print(confusion_matrix(y_test, y_pred))
    print("Classification Report:")
    print(classification_report(y_test, y_pred, zero_division=1))
    print("-" * 50)
````

# $\blacksquare$ Сравнить все обученные модели, построить их **confusion matrices**. Сделать выводы о полученных моделях в рамках решения задачи классификации на выбранных данных.

## Сравнение моделей

1. `Decision Tree и Logistic Regression` достигают идеальных результатов (точность = 1), но есть вероятность переобучения для Decision Tree, особенно на малых выборках.
2. `Random Forest` показывает более реалистичную производительность с меньшей точностью, но сохраняет устойчивость и избегает полного переобучения. Однако проблемы с классификацией редких классов (3 и 5) остаются.
3. `KNN`демонстрирует наименьшую точность среди всех моделей, но может быть полезным для сценариев с большей выборкой и оптимизацией параметра kk.
- Модели `Decision Tree и Logistic Regression` идеально справляются с <u>текущими</u> данными, но стоит протестировать их на большей выборке, чтобы проверить их "общность".

## Построение confusion matrix для каждой модели


````python
from sklearn.metrics import confusion_matrix

fig, axes = plt.subplots(2, 2, figsize=(18, 12))
axes = axes.ravel()  # Плоский массив для удобства

for i, (name, model) in enumerate(models_no_scaled.items()):
    # Получаем предсказания
    y_pred = model.predict(X_test)

    # Строим confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    # Визуализируем с помощью heatmap
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_, ax=axes[i])

    # Заголовок для каждой матрицы
    axes[i].set_title(f'{name} Confusion Matrix')
    axes[i].set_xlabel('Predicted')
    axes[i].set_ylabel('True')

# Подгонка для компоновки
plt.tight_layout()
plt.show()
````

Как было сказано выше, в сравнении, `DT` и `LR` показывают лучшую точность предсказаний, однако для оставшихся моделей есть неточности с определением редких классов 3 и 5(у `RF` ---1 ошибка, у `KNN` --- 2)

### P.S. к  коментарию, про компоненты

построим график, который показывал бы, как разное количество признаков объясняет разное количество информации в данных. Это можно сделать с помощью explained variance ratio в PCA.

````python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Применяем PCA для всех компонентов
pca = PCA()
pca.fit(X_test)

# Строим график объяснённой дисперсии
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1),
         np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
plt.title('Explained Variance vs. Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
````

то есть первая компонента обьясняет ~65% дисперсии, вторая ~82%, третья ~88%. Надо оставлять наименьшее число компонент, приближенное к 90%. Оставим 3 компоненты(между 3 и 4 не такой большой разрыв, но 3 возможно визуализировать)
